````python
current_url = "http://admoblkaluga.ru/sub/government/activity/protocol.php"
base = "http://admoblkaluga.ru"
````

````python
response = requests.get(current_url)
soup = BeautifulSoup(response.text, "html.parser")
````

````python
pages = []
for link in soup.findAll('a'):
    if 'повестка' in link.text.lower():
        href = link['href']
        if not href.startswith('http'):
            href = base + href
        response = requests.get(href)
        pages.append((href, response.text))
        print(href)
````

````python
len(pages)
````

````python
s = BeautifulSoup(pages[0][1], 'html.parser')
````

````python
seen = set()
for page in pages:
    s = BeautifulSoup(page[1], 'html.parser')
    for link in s.findAll('a'):
        if 'скачать' in link.text.lower():
            href = link['href']
            if not href.startswith('http'):
                href = base + href
            if href not in seen:
                match = re.search(r'event3=[a-zA-Z]+([0-9_]+)\.rtf', href)
                if match:
                    filename = '{}_agenda.rtf'.format(match[1])
                    urllib.request.urlretrieve(href, filename)
                    print(filename)
                    seen.add(href)
````

````python
def download_agenda(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    pages = []
    for link in soup.findAll('a'):
        if 'повестка' in link.text.lower():
            href = link['href']
            if not href.startswith('http'):
                href = base + href
            response = requests.get(href)
            pages.append((href, response.text))
            print(href)
            
    seen = set()
    for page in pages:
        s = BeautifulSoup(page[1], 'html.parser')
        for link in s.findAll('a'):
            if 'скачать' in link.text.lower():
                href = link['href']
                if not href.startswith('http'):
                    href = base + href
                if href not in seen:
                    match = re.search(r'event3=[a-zA-Z]+([0-9_]+)\.rtf', href)
                    if match:
                        filename = '{}_agenda.rtf'.format(match[1])
                        urllib.request.urlretrieve(href, filename)
                        print(filename)
                        seen.add(href)
````
