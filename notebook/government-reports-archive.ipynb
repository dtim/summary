```python
import re
import time
import pprint
import requests
import urllib.request as rq
from bs4 import BeautifulSoup
```

```python
BASE_URL = "http://admoblkaluga.ru"
ARCHIVE_URL = BASE_URL + "/sub/government/activity/archive/"
```

```python
archive_index_response = requests.get(ARCHIVE_URL)
archive_soup = BeautifulSoup(archive_index_response.text, "html.parser")
```

```python
page_regex = re.compile(r'''/sub/government/activity/archive/([0-9a-z_]+)\.php''')
indices = []
for link in archive_soup.findAll('a'):
    url = link.get('href')
    if url:
        match = re.match(page_regex, url)
        if match:
            indices.append((url, match.group(1)))
pprint.pprint(indices)
```

```python
for index, directory in indices:
    index_url = BASE_URL + index
    reply = requests.get(index_url)
    soup = BeautifulSoup(reply.text, "html.parser")
    for link in soup.find_all('a'):
        text = link.string
        href = link.get('href')
        if href and href.startswith('/upload/_meropr/'):
            url = BASE_URL + href
            filename = href.split('/')[-1]
            print(url, filename)
            rq.urlretrieve(url, filename)
            time.sleep(1)
```
